```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

DATADIR = "specs/_test"
CATEGORIES = ["classical","hiphop","jazz","metal","pop","raggae"]

for category in CATEGORIES:
    path = os.path.join(DATADIR, category) #path to different musics
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(img_array, cmap="gray")
        plt.show()
    
```
